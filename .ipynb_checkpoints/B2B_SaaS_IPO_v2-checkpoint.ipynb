# [Analysis II] Burn efficiency 

burn_ratio = [(total capital raised) - (cash on BS at IPO)] / (Revenue at IPO)

A lower burn ratio is better because it means the company spent less money to get to $100M ARR

The burn ratio can be negative. It means the company got to $100M and generated enough free cash flow so that the cash on the balance sheet exceeded the amount raised (i.e net capital raised was negative)


Observation: 
>Bottom up companies have lower median burn-rates. That is they burn less dollars to get to the same amount of ARR (or revenue). And the difference is substantial. [Talk about how much and how that translates to dollars invested]


```python
f, axes = plt.subplots(1, 1, sharey=True, figsize=(6, 6))
f.subplots_adjust(wspace=0.5)

saas_box_plot(df,'arr_ef','','Burn rate',axes,plot_type='box',axis='left')
#saas_box_plot(df,'arr_ef','','Burn rate',axes[1],plot_type='box',axis='right')
f.tight_layout()
f.text(x = 0.001, y = 1.1, s = "Bottom-up SaaS companies are more efficient",
               fontsize = 20, weight = 'bold', alpha = .75,color='black')
f.text(x = 0.03, y = 1.05, s = "Bottom-up SaaS companies tend to spend less to generate revenue",
               fontsize = 16, alpha = .85)
f.text(x = 0.2, y = 1.015, s=  "(Burn rate = capital spent to generate $100 in ARR)",
               fontsize = 12, alpha = .85)

plt.show()
#describe_col_by_type(df,'rev_ef')
describe_col_by_type(df,'arr_ef')

```

# High prevalance of Bottom-up SaaS companies in upper quartiles


```python
def saas_stacked_chart(df,param,q=4):
    df_t = df
    with ChainedAssignent():
        df_t['q']= pd.qcut(df_t[param],q)

    labels = list()
    for i in range(q):
        labels.append(str(i))


    bot = np.array(df_t[df_t.type=='bot_up_ent'].groupby('q')[param].count())
    top = np.array(df_t[df_t.type=='top_down_ent'].groupby('q')[param].count())
    bot_f = bot/(bot+top)*100
    top_f = top/(bot+top)*100

    # Data
    x=['0-25','25-50','50-75','75-100']
    y=[]
    y.append(bot_f)
    y.append(top_f)
    return (x,y)
    
```

```python
f, axes = plt.subplots(1, 3, sharey=False, figsize=(12,5))
f.subplots_adjust(wspace=0.5)
x,y = saas_stacked_chart(df,'arr_ef')
axes[0].stackplot(x,y,labels=['Bottom-up','Top-down'])
axes[0].legend(loc='upper left')

x,y = saas_stacked_chart(df,'ltm_median_payback_period')
axes[1].stackplot(x,y,labels=['Bottom-up','Top-down'])
axes[1].legend(loc='upper left')

x,y = saas_stacked_chart(df,'net_dollar_retention')
axes[2].stackplot(x,y,labels=['Bottom-up','Top-down'])
axes[2].legend(loc='upper left')



plt.show()
```

# Cluster analysis


```python
def numerify_type(df):
    df_km = df
    df_km['num_type']=0
    df_c_b = df_km[df_km.type == 'bot_up_ent']
    df_c_t = df_km[df_km.type == 'top_down_ent']
    df_c_t['num_type']=1
    df_km = pd.concat([df_c_b,df_c_t],axis=0)
    return df_km

from sklearn import metrics
from sklearn.cluster import KMeans

filter_cols = ['arr_ef', 'arr_growth']
df_c0 = df[df.arr_ef<df.arr_ef.quantile(0.25)]
l = len(filter_cols)
df_c = df_c0[filter_cols]
for c in filter_cols:
    df_c = df_c[df_c[c].notna()]

if ('type' in filter_cols):
    df_c = numerify_type(df_c)
    df_c = df_c.drop('type', 1)

x = df_c.iloc[:, list(range(l))].values
clusters=2
kmeans = KMeans(n_clusters = clusters) 
y = kmeans.fit_predict(x) 
  
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```

```python
se = pd.Series(kmeans.labels_)
df_c['new_col'] = se.values
df_c0['new_col'] = df_c['new_col']
```

```python
df_c0[(df_c0.type == 'top_down_ent') & (df_c0.new_col == 0)][['ltm_median_payback_period','net_dollar_retention','type','new_col']]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x[:,0],x[:,1],c=y,cmap='rainbow')
```
