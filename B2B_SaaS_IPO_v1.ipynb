```python
def empirical_cdf(data):
    percentiles = []
    n = len(data)
    sort_data = np.sort(data)
  
    for i in np.arange(1,n+1):
        p = i/n
        percentiles.append(p)
    return sort_data,percentiles

```

```python
param = 'rev_ef'
df_t = df_ex_smb[df_ex_smb[param].notna()]

fig,ax = plt.subplots()

bot_sort,bot_ptile = empirical_cdf(df_t[df_t.type=='bot_up_ent'][param])
ax.plot(bot_sort,bot_ptile, label='Data',color = 'blue')
plt.title(r'$\bf{Empirical \ CDF}$')
plt.xlabel(param)
plt.ylabel('percentiles')

top_sort,top_ptile = empirical_cdf(df_t[df_t.type=='top_down_ent'][param])



ax.plot(top_sort,top_ptile, label='Data',color = 'green')

df_bt = df_t[df_t.type == 'bot_up_ent']
ymedian = np.empty(len(df_bt[param]))
ymedian.fill(0.5)
ymax = np.empty(len(df_bt[param]))
ymax.fill(np.max(bot_ptile))
ymin = np.empty(len(df_bt[param]))
ymin.fill(np.min(bot_ptile))
y25 = np.empty(len(df_bt[param]))
y25.fill(0.25)
y75 = np.empty(len(df_bt[param]))
y75.fill(0.75)

# Plot the lines
max_line = ax.plot(bot_sort,ymax, label='Max', linestyle='--',color = '#001c58')
y75_line = ax.plot(bot_sort,y75, label='75Percentile', linestyle='--',color = '#ff7a69')
med_line = ax.plot(bot_sort,ymedian, label='Median', linestyle='--',color = '#fb6500')
y25_line = ax.plot(bot_sort,y25, label='25Percentile', linestyle='--',color = '#4fb4b1')
min_line = ax.plot(bot_sort,ymin, label='Min', linestyle='--',color = '#ffbb39')

#annotate lines
ax.annotate('Max/$100^{th}$ Percentile',xy = (1,0.96),color = '#001c58')
ax.annotate('$75^{th}$ Percentile',xy = (1,0.71),color = '#ff7a69')
ax.annotate('Median/$50^{th}$ Percentile',xy = (1,0.46),color = '#fb6500')
ax.annotate('$25^{th}$ Percentile',xy = (1,0.21),color = '#4fb4b1')
ax.annotate('Min/$1^{st}$ Percentile',xy = (1.3,0.04),color = '#ffbb39')
#final and important bits
plt.tight_layout()
plt.title(r'$\bf{Empirical \ CDF}$')
plt.ylabel('percentiles')


plt.show()

```
